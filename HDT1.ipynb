Lectura de csv

```python
with open('Admission_Predict.csv') as csv_file:
    datos = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in datos:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        else:
            print(f'\t{row[0]} Puntaje GRE {row[1]}, Puntaje TOEFL {row[2]}.')
            line_count += 1
    print(f'Processed {line_count} lines.')
```
