# Se tomaran las columas de GRE Score, TOEFL Score y CGPA 

```python
def gradient_descent(
        X,
        y,
        theta_0,
        cost,
        cost_derivate,
        alpha=0.01,
        treshold=0.0001,
        max_iter=10000, 
        lamb=1):
    theta, i = theta_0, 0
    costs = []
    gradient_norms = []
    while np.linalg.norm(cost_derivate(X, y, theta, lamb)) > treshold and i < max_iter:
        theta -= alpha * cost_derivate(X, y, theta, lamb)
        i += 1
        costs.append(cost(X, y, theta, lamb))
        gradient_norms.append(cost_derivate(X, y, theta, lamb))
    return theta, costs, gradient_norms
```

```python
def linear_cost_derivate(X, y, theta):
    h = np.matmul(X, theta)
    m, _ = X.shape
    return np.matmul((h - y).T, X).T / m
```

```python
def linear_cost_derivate_regular(X, y, theta, lamb):
    h = np.matmul(X, theta)
    m, _ = X.shape
    return np.matmul(((h - y).T, X).T + lamb * h) / m
```

```python
def linear_cost(X, y, theta):
    m, _ = X.shape
    h = np.matmul(X, theta)
    sq = (y - h) ** 2
    return sq.sum() / (2 * m)
```

```python
def linear_cost_regular(X, y, theta, lamb):
    m, _ = X.shape
    h = np.matmul(X, theta)
    sq = (y - h) ** 2
    jtheta = h**2
    return sq.sum() + lamb*jtheat.sum() / (2 * m)
```
